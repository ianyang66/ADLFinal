```python
import pandas as pd
```

```python
users = pd.read_csv('../data/users.csv').fillna('')
users
```

```python
group = []
subgroup = []
for interest in users['interests']:
    gp = []
    sub = []
    for i in interest.split(','):
        try:
            g, s = i.split('_')
        except:
            continue
        if(g not in gp):
            gp.append(g)
        if(s not in sub):
            sub.append(s)
    group.append(','.join(gp))
    subgroup.append(','.join(sub))
users['group'] = group
users['subgroup'] = subgroup
users
    
```

```python
courses = pd.read_csv('../data/courses.csv')
print(courses['groups'].isna().sum())
courses = courses.fillna('')
```

```python
import itertools
print(list(itertools.combinations('123', 2)))
```

```python
def get_combinations(comb):
    t = []
    for i in range(1, 6):
        t += [','.join(sorted(list(x))) for x in list(itertools.combinations(comb, i))]
    return t[::-1]
```

```python
course_by_group ={}
for id, groups in zip(courses['course_id'][:], courses['groups'][:]):
    if(groups == ''):
        continue
    gp_split = get_combinations(groups.split(','))
    #print(gp_split)
    for g in gp_split:
        try:
            course_by_group[g].append(id)
        except:
            course_by_group[g] = [id]
    
course_by_group.keys()
```

```python
train = pd.read_csv('../data/train.csv')
bought_courses = {}
for id in train['course_id']:
    for i in id.split(' '):
        try:
            bought_courses[i] += 1
        except:
            bought_courses[i] = 1
    
bought_times = sorted(list(bought_courses.items()), key = lambda x:x[1], reverse=True)
garbage_course = list(set(courses['course_id']).difference(set(bought_courses.keys())))
popular_course_id = [x[0] for x in bought_times] + garbage_course
#len(popular_course_id)
#len(garbage_course)
#len(popular_course_id)
bought_times[:20]
```

```python
user_interest_group = {}
for id, group in zip(users['user_id'], users['group']):
    user_interest_group[id] = group.split(',')
list(user_interest_group.items())[:10]
```

```python
test_unseen = pd.read_csv('../data/test_unseen.csv')
top_k = 50
answer = []
for id in test_unseen['user_id'][:]:
    group_index = get_combinations((user_interest_group[id]))
    #print(group_index)
    rec = []
    group_len = 6 #len(group_index[0].split(',')) + 1
    temp = [] #[x[0] for x in bought_times[:9]]
    for gps in group_index:
        if(group_len > len(gps.split(','))):
           group_len = len(gps.split(','))
           rec.extend(sorted(temp, key = popular_course_id.index))
           temp = []          
        try:
            new_course =list(set(course_by_group[gps]).difference(rec))
            #print(new_course)
            temp.extend(new_course)
                
        except:
            continue
    rec.extend(sorted(temp, key = popular_course_id.index))
    not_enough = max(top_k - len(rec), 0)
    x = 0
    while(len(rec) < top_k):
        if(bought_times[x][0] not in rec):
            rec.append(bought_times[x][0])
        x += 1

    answer.append([id, sorted(rec[:top_k], key = popular_course_id.index)])
```

```python
with open('./submit_unseen_course.csv', 'w') as f:
    f.write('user_id,course_id\n')
    for id, recommendation in answer:
        f.write('{},{}\n'.format(id, ' '.join(recommendation)))
```
