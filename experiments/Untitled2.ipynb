```python
train = pd.read_csv('../data/train.csv')
train
```

```python
bought_times = {}
for co in train['course_id'].to_list():
    for c in co.split(' '):
        try:
            bought_times[c] += 1
        except:
            bought_times[c] = 1
popular = sorted(list(bought_times.items()), key=lambda x:x[1], reverse=True)
df2 = pd.DataFrame(popular, columns=['course_id', 'bought_times'])
df3 = pd.merge(df2, courses, on=['course_id'])
df3
```

```python
popular_string = ' '.join([x[0] for x in popular])
```
