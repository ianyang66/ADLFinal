```python
import pandas as pd
```

```python
users = pd.read_csv('../data/users.csv').fillna('')
users
```

```python
group = []
subgroup = []
for interest in users['interests']:
    gp = []
    sub = []
    for i in interest.split(','):
        try:
            g, s = i.split('_')
        except:
            continue
        if(g not in gp):
            gp.append(g)
        if(s not in sub):
            sub.append(s)
    group.append(','.join(gp))
    subgroup.append(','.join(sub))
users['group'] = group
users['subgroup'] = subgroup
users
    
```
