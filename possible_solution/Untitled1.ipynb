```python
import pandas as pd
import json
```

```python
df = pd.read_csv('../data/users.csv').fillna('')
df
```

```python
courses_df = pd.read_csv('../data/courses.csv').fillna('')
subgroup_df = pd.read_csv('../data/subgroups.csv').fillna('')
courses_df
```

```python
subgroup_df
```

```python
subgroup_dict = {}
for i, sub_name in enumerate(subgroup_df['subgroup_name']):
    subgroup_dict[sub_name] = i
subgroup_dict
```

```python
course_onehot_dict = {}
for id, subgroup in zip(courses_df['course_id'], courses_df['sub_groups']):
    try:
        t = [subgroup_dict[x] for x in subgroup.split(',')]
        course_onehot_dict[id] = t
    except:
        course_onehot_dict[id] = []
course_onehot_dict
```

```python
with open('../data/course_subgroup_idx.json', 'w') as f:
    json.dump(course_onehot_dict, f)
```
