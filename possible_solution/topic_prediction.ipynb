```python
import pandas as pd
import random
top_k = 50
category = 'unseen'
```

```python
df = pd.read_csv('../data/users.csv')
df = df.drop(['gender', 'occupation_titles', 'recreation_names'], axis=1).fillna("")
df
```

```python
d1 = {}
for i in df['interests'].to_list():
    for j in i.split(','):
        try:
            d1[j] += 1
        except:
            d1[j] = 1
d1 = sorted(list(d1.items()), key = lambda x:x[1], reverse=True)
d1 = [x[0] for x in d1 if x[0] != '']
d1
```

```python
df2 = pd.read_csv('../data/test_'+ category+'_group.csv')
df2
```

```python
df3 = pd.read_csv('../data/train_group.csv').fillna('')
df3
```

```python
merged = df2.merge(df, on=['user_id']).drop('subgroup', axis=1)
if(category == 'seen'):
    merged = merged.merge(df3, on=['user_id'])
merged[:20]
```

```python
courses_df = pd.read_csv('../data/courses.csv').fillna('')
couses_price_zero = courses_df[courses_df['course_price'] == 0]
couses_price_zero
```

```python
d2 = {}

for courses in courses_df['groups+subgroups'].to_list():
    for c in courses.split(','):
        try:
            d2[c] += 1
        except:
            d2[c] = 1
d2 = sorted(list(d2.items()), key=lambda x:x[1], reverse=True)
d2 = [x[0] for x in d2 if x[0] != '']
d2
```

```python
popular_sub = []
for x in d1:
    try:
        popular_sub.append([x, d1.index(x) + d2.index(x)])
    except:
        print(x)
        popular_sub.append([x, 2 * d1.index(x)])
popular_sub_lst = [x[0] for x in sorted(popular_sub, key=lambda x:x[1])]
popular_sub
```

```python
GROUPS = {  '00_空值':[0], '01_人文':[62, 71, 85], '02_手作':[9, 17, 18, 24, 27, 31, 46, 89], \
            '03_投資理財':[64, 69, 70, 76], \
            '04_攝影':[19, 40, 53, 56, 73, 86, 90], \
            '05_生活品味':[1, 2, 13, 14, 15, 16, 29, 55, 79, 82, 84], \
            '06_程式':[32, 33, 34, 41, 42, 43, 44, 52, 57, 58, 63, 68, 74, 75, 80, 81], \
            '07_職場技能':[7, 35, 37,50, 51, 54, 59, 61, 78, 83, ], \
            '08_藝術':[4, 5, 12, 20, 22, 23, 36, 60], '09_行銷':[65, 66, 72, 87, 91], \
            '10_設計':[3, 6, 25, 38, 39, 67, 77], '11_語言':[8, 28, 45, 47, 48, 88], '12_音樂':[10, 11, 21, 26, 30, 49]}
subgroup = pd.read_csv('../data/subgroups.csv')
group_dict = {}
for i in GROUPS.keys():
    for j in GROUPS[i]:
        group_dict[j] = int(i.split('_')[0])
subgroup_dict = subgroup.set_index('subgroup_name').to_dict()['subgroup_id']
subgroup_dict
```

```python
err_sub = []
for i, x in enumerate(popular_sub):
    try:
        popular_sub[i] = [x, subgroup_dict[x.split('_')[1]]]
    except:
        err_sub.append(x)
err_sub
for x in err_sub:
    popular_sub.remove(x)
popular_sub
```

```python
random.seed(55)
rec_sub = []
subgroup_set = set(subgroup_dict.values())
popular_sub_set = set(range(0, 92))
pop_sub_values = [x[1] for x in popular_sub]
if(category == 'seen'):
    for interest, sub in zip(merged['interests'], merged['subgroup']):
        temp = []
        for i in interest.split(','):

            try:
                i = i.split('_')[1]
                temp.append(subgroup_dict[i])
            except:
                continue
        temp.sort(key=pop_sub_values.index)
        usr_interest = temp + sorted(list(set([int(x) for x in sub.split(' ') if x != '']).difference((set(temp)))), key=pop_sub_values.index)
        #分三段

        n = max(0, top_k - len(usr_interest))
        added_sub = []
        for p in popular_sub:
            if n == 0:
                break
            if(p[1] in usr_interest):
                continue
            added_sub.append(p[1])
            n -= 1
        rec_sub.append(sorted(usr_interest, key=pop_sub_values.index) + sorted(added_sub, key=pop_sub_values.index))
else:
    for interest in merged['interests']:
        temp = []
        for i in interest.split(','):

            try:
                i = i.split('_')[1]
                temp.append(subgroup_dict[i])
            except:
                continue

        usr_interest = temp

        n = max(0, top_k - len(usr_interest))
        added_sub = []
        for p in popular_sub:
            if n == 0:
                break
            if(p[1] in usr_interest):
                continue
            added_sub.append(p[1])
            n -= 1
        rec_sub.append(usr_interest + sorted(added_sub, key=pop_sub_values.index))
```

```python
with open('./submit_'+ category+'.csv', 'w') as f:
    f.write('user_id,subgroup\n')
    for user, rec in zip(df2['user_id'].to_list(), rec_sub):
        f.write('{},{}\n'.format(user, ' '.join([str(x) for x in rec])))
```
