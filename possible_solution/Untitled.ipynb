```python
import pandas as pd
import random
top_k = 15
category = 'seen'
```

```python
df = pd.read_csv('../data/users.csv')
df = df.drop(['gender', 'occupation_titles', 'recreation_names'], axis=1).fillna("")
df
```

```python
d1 = {}
for i in df['interests'].to_list():
    for j in i.split(','):
        try:
            d1[j] += 1
        except:
            d1[j] = 1
d1 = sorted(list(d1.items()), key = lambda x:x[1], reverse=True)
d1 = [x[0] for x in d1][1:]
d1
```

```python
df2 = pd.read_csv('../data/test_'+ category+'_group.csv')
df2
```

```python
merged = df2.merge(df, on=['user_id']).drop('subgroup', axis=1)
merged[:20]
```

```python
courses_df = pd.read_csv('../data/courses.csv')
couses_price_zero = courses_df[courses_df['course_price'] == 0]
couses_price_zero
```

```python
GROUPS = {  '00_空值':[0], '01_人文':[62, 71, 85], '02_手作':[9, 17, 18, 24, 27, 31, 46, 89], \
            '03_投資理財':[64, 69, 70, 76], \
            '04_攝影':[19, 40, 53, 56, 73, 86, 90], \
            '05_生活品味':[1, 2, 13, 14, 15, 16, 29, 55, 79, 82, 84], \
            '06_程式':[32, 33, 34, 41, 42, 43, 44, 52, 57, 58, 63, 68, 74, 75, 80, 81], \
            '07_職場技能':[7, 35, 37,50, 51, 54, 59, 61, 78, 83, ], \
            '08_藝術':[4, 5, 12, 20, 22, 23, 36, 60], '09_行銷':[65, 66, 72, 87, 91], \
            '10_設計':[3, 6, 25, 38, 39, 67, 77], '11_語言':[8, 28, 45, 47, 48, 88], '12_音樂':[10, 11, 21, 26, 30, 49]}
subgroup = pd.read_csv('../data/subgroups.csv')
group_dict = {}
for i in GROUPS.keys():
    for j in GROUPS[i]:
        group_dict[j] = int(i.split('_')[0])
subgroup
```

```python
random.seed(55)
sub = []
for i in merged['interests']:
    a = [x for x in i.split(',') if x != '']
    
    if(len(a) > top_k):
        #samples = random.sample(a, k = 5)
        samples = sorted(a, key=d1.index)[:top_k]
        sample_choice = [subgroup.loc[subgroup['subgroup_name'] == x.split('_')[1]]['subgroup_id'] for x in samples]
        b = []
        n = 1
        for i, x in enumerate(sample_choice):
            try:
                x = x.iloc[0]
                b.append(x)
            except:
                x = d1[5 + n]
                n += 1
            sample_choice = b
    else:
        if(len(a) == 0):
            sample_choice = d1[:top_k]
            sample_choice = [subgroup.loc[subgroup['subgroup_name'] == x.split('_')[1]]['subgroup_id'] for x in samples]
        else:
            sample_choice = a
            sample_choice.extend(d1[:top_k - len(a)])
            sample_choice = [subgroup.loc[subgroup['subgroup_name'] == x.split('_')[1]]['subgroup_id'] for x in sample_choice]
            n = 0
            b = []
            #print(sample_choice)
            for i, x in enumerate(sample_choice):
                try:
                    x = x.iloc[0]
                    b.append(x)
                except:
                    x = d1[n]
                    n += 1
                    b.append(subgroup.loc[subgroup['subgroup_name'] == x.split('_')[1]]['subgroup_id'].iloc[0])
            sample_choice = b
            
    
    sub.append(sample_choice)
#merged['interests'][:20]
sub[:20]
```

```python
df3 = pd.DataFrame(sub)
df3['new_col'] = df3.apply(lambda x: ' '.join(x.astype(str).values), axis=1)
df3
```

```python
merged2 = pd.concat([merged['user_id'], df3['new_col']], axis = 1)
merged2.rename(columns={'new_col': 'subgroup'}, inplace=True)
merged2
```

```python
merged2.to_csv('./submit_'+ category+'.csv',  index=False)
```
